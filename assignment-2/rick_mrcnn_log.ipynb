## Plot logs obtaineds during train 

```python
import matplotlib.pyplot as plt
```

* rpn_class_loss = RPN anchor classifier loss
* rpn_bbox_loss = RPN bounding box loss graph
* mrcnn_class_loss = loss for the classifier head of Mask R-CNN
* mrcnn_bbox_loss = loss for Mask R-CNN bounding box refinement
* mrcnn_mask_loss = mask binary cross-entropy loss for the masks head

```python
train_loss = list()
train_rpn_class = list()
train_rpn_bbox_loss = list()
train_mrcnn_class_loss = list()
train_mrcnn_bbox_loss = list()
train_mrcnn_mask_loss = list()

val_loss = list()
val_rpn_class = list()
val_rpn_bbox_loss = list()
val_mrcnn_class_loss = list()
val_mrcnn_bbox_loss = list()
val_mrcnn_mask_loss = list()

with open('train-log.txt') as f:
    for line in f:
        line = line.split()
        if len(line) > 2:
            # train logs
            train_loss.append(float(line[7]))
            train_mrcnn_class_loss.append(float(line[16]))
            train_mrcnn_bbox_loss.append(float(line[19]))
            train_mrcnn_mask_loss.append(float(line[22]))
            
            # val logs
            val_loss.append(float(line[25]))
            val_mrcnn_class_loss.append(float(line[34]))
            val_mrcnn_bbox_loss.append(float(line[37]))
            val_mrcnn_mask_loss.append(float(line[40]))
```

```python
plt.plot(range(len(train_mrcnn_class_loss)), train_mrcnn_class_loss, label='Train')
plt.plot(range(len(val_mrcnn_class_loss)), val_mrcnn_class_loss, label='Validation')
plt.title('Mask-RCNN - Class loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
plt.plot(range(len(train_mrcnn_bbox_loss)), train_mrcnn_bbox_loss, label='Train')
plt.plot(range(len(val_mrcnn_bbox_loss)), val_mrcnn_bbox_loss, label='Validation')
plt.title('Mask-RCNN - Bbox loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
plt.plot(range(len(train_mrcnn_mask_loss)), train_mrcnn_mask_loss, label='Train')
plt.plot(range(len(val_mrcnn_mask_loss)), val_mrcnn_mask_loss, label='Validation')
plt.title('Mask-RCNN - Mask loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
