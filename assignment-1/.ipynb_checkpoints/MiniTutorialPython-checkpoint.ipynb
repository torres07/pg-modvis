#### Numpy...

https://docs.scipy.org/doc/numpy-1.13.0/user/quickstart.html

##### Operações básicas

```python
a = np.array([[1, 1, 1], [2, 2, 2]])
b = np.array([[3, 3, 3], [3, 3, 3]])
```

```python
a[0]
```

```python
a.shape
```

```python
a.size
```

```python
# Adição

a + b
```

```python
# Multiplicação elemento a elemento

a*b
```

```python
## Atenção: o operador * não serve para multiplicação matricial!

a*np.transpose(b)
```

```python
# Multiplicação matricial

np.dot(a, np.transpose(b))
```

```python
# Exponenciação

print(a)
np.power(a, 3)
```

```python
# Somar todos os elementos do array

np.sum(a)
```

```python
# Fazer a média dos elementos do array

np.mean(a)
```

##### 3D

https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

ax.set_xlim3d((0, 1))
ax.set_ylim3d((0, 1))
ax.set_zlim3d((0, 1))

points = np.random.random((50, 3))

for i in points:
    ax.scatter(i[0], i[1], i[2], c = 'r', marker = ',')
    
plt.show()
```
